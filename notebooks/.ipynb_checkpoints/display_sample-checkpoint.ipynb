```python
import os
from pathlib import Path
from typing import Tuple

import matplotlib.pyplot as plt
import torch
import torchvision

from src.data.load_data import OmniglotLoader
from src.data.sampling import extract_sample
```

```python
def display_sample(sample: torch.Tensor, figsize: Tuple[int, int] = (16, 7)) -> None:
    """
    Displays a few-shot learning sample in a grid layout.

    :param sample: Tensor of images to display with shape (n_way, n_examples, C, H, W)
    :param figsize: Size of the matplotlib figure (width, height) in inches
    :return: None
    """
    # Validate input tensor dimensions
    if sample.dim() != 5:
        raise ValueError(f"Expected 5D tensor (n_way, n_examples, C, H, W), got {sample.dim()}D tensor")
    if sample.shape[2] not in [1, 3]:
        raise ValueError(f"Expected 1 or 3 channels, got {sample.shape[2]} channels")

    try:
        # Convert 5D tensor to 4D by flattening first two dimensions
        n_way, n_examples = sample.shape[:2]
        sample_4d = sample.view(n_way * n_examples, *sample.shape[2:])

        # Create image grid
        grid = torchvision.utils.make_grid(sample_4d, nrow=n_examples)

        # Convert CHW to HWC for matplotlib
        grid_np = grid.permute(1, 2, 0).cpu().numpy()

        # Normalize if needed
        if grid_np.max() > 1:
            grid_np = grid_np / 255.0

        # Display the grid
        plt.figure(figsize=figsize)
        plt.imshow(grid_np)
        plt.axis("off")
        plt.tight_layout()
        plt.show()

    except Exception as e:
        raise RuntimeError(f"Failed to display sample: {str(e)}")
```
